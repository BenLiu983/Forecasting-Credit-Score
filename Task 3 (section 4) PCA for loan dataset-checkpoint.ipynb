```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
%matplotlib inline
```

Import loan dataset

```python
data = pd.read_csv("loan_final313.csv")
```

```python
df = pd.DataFrame(data)
display(df.head())
display(df.shape)
```

#  PCA1 (raw data)

```python
df1 = df.iloc[:,[8, 4, 21, 10, 12, 20, 6, 16, 18, 23]]
display(df1.head(20))
display(df1.shape)
```

```python
X1 = df1.iloc[:,:-1]
display(X1.head())
```

```python
y1 = df1.iloc[:,-1]
y1.head()
```

```python
# Create instances 
scaler = StandardScaler()
# Apply scaler
X1 = scaler.fit_transform(X1)
display(type(X1))
X1 = pd.DataFrame(X1)
display(type(X1))
```

```python
# Create instances 
pca1 = PCA(n_components=2)
# Create a fitting model
pca1_model = pca1.fit(X1)
# Convert to a dataframe
X1_trans = pd.DataFrame(pca1_model.transform(X1), columns=['pca1', 'pca2'])
display(X1_trans.head())
```

```python
#plt.scatter(X1_trans['pca1'], X1_trans['pca2'], c=y1, s=1)
X1_trans['y_values']=y1
display(X1_trans['y_values'].shape)
X1_trans=X1_trans[X1_trans['pca2']<7.5]
display(X1_trans.shape)
display(X1_trans['y_values'].shape)
plt.scatter(X1_trans['pca1'], X1_trans['pca2'], c=X1_trans['y_values'], s=0.2)
plt.title('PCA for raw dataset after scaling (y=grade)')
```

```python
comp1 = pd.DataFrame(pca1_model.components_, columns=X1.columns)
display(comp1)
```

# PCA2 (clean data and leave 6.05 as it is)

Data cleansing

```python
# Exclude "other", "none", "any" from the home ownership category
df2 = df[df['home_ownership_cat']<4]
display(df2.shape)
```

```python
# Exclude "Joint" from the application type
df2 = df2[df2['application_type_cat']==1]
display(df2.shape)
```

```python
# Only remain "credit card", "other", "debt consolidation", "home improvement", "major purpose" in the purpose
some = ['credit_card', 'other', 'debt_consolidation', 'home_improvement', 'major_purchase']
df2 = df2.loc[df2['purpose'].isin(some)] 
display(df2.shape)
```

Select 10 important features

```python
df2 = df2.iloc[:,[8, 4, 21, 10, 12, 20, 6, 16, 18, 23]]
display(df2.head())
display(df2.shape)
```

Construct X and y (1 raw dataset and 2 clean datasets)

```python
X2 = df2.iloc[:,:-1]
display(X2.head())
y2 = df2.iloc[:,-1]
display(y2.head())
```

```python
# Create instances 
scaler = StandardScaler()
# Apply scaler
X2 = scaler.fit_transform(X2)
display(type(X2))
X2 = pd.DataFrame(X2)
display(type(X2))
display(X2.head())
```

```python
# Create instances 
pca2 = PCA(n_components=2)
# Create a fitting model
pca2_model = pca2.fit(X2)
# Convert to a dataframe
X2_trans = pd.DataFrame(pca2_model.transform(X2), columns=['pca1', 'pca2'])
display(X2_trans.head())
```

Scatterlot

```python
#plt.scatter(X2_trans['pca1'], X2_trans['pca2'], c=y2, s=1)
X2_trans['y_values']=y2
display(X2_trans['y_values'].shape)
X2_trans=X2_trans[X2_trans['pca2']<7.5]
display(X2_trans.shape)
display(X2_trans['y_values'].shape)
plt.scatter(X2_trans['pca1'], X2_trans['pca2'], c=X2_trans['y_values'], s=0.2)
plt.title('PCA for clean dataset (contain 6.05) after scaling (y=grade)')
```

Explained variance of PCA components

```python
comp2 = pd.DataFrame(pca2_model.components_, columns=X2.columns)
display(comp2)
```

Annual income and loan amount take up large percent of pca1 and pca2, which means they explain major variation of the dataset, mainly because of their large scale, so we try to scale the dataset before conducting PCA.
