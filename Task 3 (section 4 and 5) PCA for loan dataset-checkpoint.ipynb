```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
%matplotlib inline
```

Import loan dataset

```python
data = pd.read_csv("loan_final313.csv")
```

```python
df = pd.DataFrame(data)
display(df.head())
display(df.shape)
```

Data cleansing

```python
# Exclude "other", "none", "any" from the home ownership category
data0 = df[df['home_ownership_cat']<4]
display(data0.shape)
```

```python
# Exclude "Joint" from the application type
data0 = data0[data0['application_type_cat']==1]
display(data0.shape)
```

```python
# Only remain "credit card", "other", "debt consolidation", "home improvement", "major purpose" in the purpose
some = ['credit_card', 'other', 'debt_consolidation', 'home_improvement', 'major_purchase']
data0 = data0.loc[data0['purpose'].isin(some)] 
display(data0.shape)
```

```python
# prepare for cleaning type 2
data1 = data0
data1 = data1[data1['emp_length_int']!=6.05]
display(data1.shape)
```

```python
# Round 6.05 to 6 in ecployment length(years)
data0[data0['emp_length_int']==6.05]=6
display(data0.shape)
display(data0[data0['emp_length_int']==6.05].shape)
```

Select 10 important features

```python
df1 = df.iloc[:,[8, 4, 21, 10, 12, 20, 6, 16, 18, 23]]
display(df1.head())
display(df1.shape)
```

```python
df1_clean = data0.iloc[:,[8, 4, 21, 10, 12, 20, 6, 16, 18, 23]]
display(df1_clean.head())
display(df1_clean.shape)
```

```python
df1_clean2 = data1.iloc[:,[8, 4, 21, 10, 12, 20, 6, 16, 18, 23]]
display(df1_clean2.head())
display(df1_clean2.shape)
```

# PCA 1 ( "grade" as y, before scaling)

Construct X and y (1 raw dataset and 2 clean datasets)

```python
X1 = df1.iloc[:,:-1]
display(X1.head())
```

```python
y1 = df1.iloc[:,-1]
y1.head()
```

PCA with 2 components (good for visualization)

```python
# Create instances 
pca1 = PCA(n_components=2)
# Create a fitting model
pca1_model = pca1.fit(X1)
# Convert to a dataframe
X1_trans = pd.DataFrame(pca1_model.transform(X1), columns=['pca1', 'pca2'])
display(X1_trans.head())
```

```python
# Create instances 
pca1_clean = PCA(n_components=2)
# Create a fitting model
pca1_clean_model = pca1_clean.fit(X1_clean)
# Convert to a dataframe
X1_clean_trans = pd.DataFrame(pca1_clean_model.transform(X1_clean), columns=['pca1', 'pca2'])
display(X1_clean_trans.head())
```

```python
# Create instances 
pca1_clean2 = PCA(n_components=2)
# Create a fitting model
pca1_clean2_model = pca1_clean2.fit(X1_clean2)
# Convert to a dataframe
X1_clean2_trans = pd.DataFrame(pca1_clean2_model.transform(X1_clean2), columns=['pca1', 'pca2'])
display(X1_clean2_trans.head())
```

Scatterlot

```python
plt.scatter(X1_trans['pca1'], X1_trans['pca2'], c=y1, s=1)
plt.title('PCA for raw dataset before scaling (y=grade)')
```

```python
plt.scatter(X1_clean_trans['pca1'], X1_clean_trans['pca2'], c=y1_clean, s=1)
plt.title('PCA for clean dataset (type 1) before scaling (y=grade)')
```

```python
plt.scatter(X1_clean2_trans['pca1'], X1_clean2_trans['pca2'], c=y1_clean2, s=1)
plt.title('PCA for clean dataset (type 2) before scaling (y=grade)')
```

There are many overlaps, it seems like it is difficult to classify.

Explained variance of PCA components

```python
comp1 = pd.DataFrame(pca1_model.components_, columns=X1.columns)
display(comp1)
```

```python
comp1_clean2 = pd.DataFrame(pca1_clean2_model.components_, columns=X1_clean2.columns)
display(comp1_clean2)
```

5 principal components can explain almost 80% of the variation of the overall dataset.
